# Basic GPA calculator
***
The following code presents my first ever Python project from my college days. This mini project boasts lack of coding knowledge at the time of production, as you would find the following drawbacks:
* Only capital letter entries for grades work in this calculator.
* There are no prompts for users to re-input values after entering invalid values.
* Following the point above, you would have to start the calculator again and input correct values.
* Many more issues that real programmers can point out and roast me with.

After gaining a little more coding knowledge, you will find these issues fixed in my **GPA 2.0** code :)

```python
Total_grades = 0.0 #Total grade value counts
Total_credits = 0.0 #Total credit hour counts

#Inputting the number of courses to be used for our GPA calculation
courses = int(input("Please input the number of courses: "))  

#A note for users that only the following grade letters are valid for this GPA calculator
print("\nValid grade letters for this calculator are: A+, A, B+, B, C+, C, D, F\n")

#While loop is used for the bulk of our calculation
index = 1
while index<courses+1: 
    #Looping index of courses with respective grade entries
    print("Course #",index,", Grade:", end=' ') 
    grade = input()
    
    a1 = "A+" #Assigning variables for our grades to be used for calculation
    a2 = "A"
    b1 = "B+"
    b2 = "B"
    c1 = "C+"
    c2 = "C"
    d = "D"
    f = "F"
    
    if grade == a1:    #Conditional statements are applied for each grade,
        gv = 4.5       #where a grade value is returned per specific grade.
    elif grade == a2:  #Any value entered except for the following grades would 
        gv = 4.0       #notify the user their input is invalid.
    elif grade == b1:
        gv = 3.5
    elif grade == b2:
        gv = 3.0
    elif grade == c1:
        gv = 2.5
    elif grade == c2:
        gv = 2.0
    elif grade == d:
        gv = 1.0
    elif grade == f:
        gv = 0.0
    else:
        print("This is invalid, please input grades like: A+,B, or C- etc..Reset Calculator")
        break
    
    #Input for course credits per course, any value besides 3 or 4 
    #in this case would tell the user it's invalid.
    print("Course credits (3 OR 4): ", end=' ')
    credit = int(input())
    print("")
    
    if credit<3 or credit>4:
        print("Invalid value, input either 3 OR 4. Reset Calculator.")
        break
    elif credit>3 and credit<4:
        print("Invalid value, input either 3 OR 4. Reset Calculator.")
        break
    
    #Main calculations:
    GCP = gv*credit #Multiplying grade values and credits for each course
    Total_grades += GCP #The product of the Grade-values and credit hours are summed
    Total_credits += credit #Adding up all course credit hours
    
    index+=1 #Finishing our while loop with increment
    
#Finally the last part of our calculation, dividing the total
#grade_value-credit_hour products by the total number of credit hours
GPA = Total_grades/Total_credits

########################################################Getting our GPA calculator output!
print("The total GPA of your",courses,"courses is ", GPA)
```
